```python
import pandas as pd
import re
import string
import os
```

```python
# Path ke data mentah
raw_path = "../data/raw/GojekReview.csv"

# Baca file CSV
df = pd.read_csv(raw_path)

print("Jumlah data:", len(df))
df.head()
```

```python
def clean_text(text):
    text = str(text).lower()  # lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # hapus URL
    text = re.sub(r'\@\w+|\#', '', text)  # hapus mention & hashtag
    text = re.sub(r'\d+', '', text)  # hapus angka
    text = text.translate(str.maketrans('', '', string.punctuation))  # hapus tanda baca
    text = re.sub('\s+', ' ', text).strip()  # hapus spasi ganda dan pinggiran
    return text
```

```python
# Terapkan ke kolom full_text
df["clean_text"] = df["full_text"].apply(clean_text)

# Tampilkan hasil
df[["full_text", "clean_text"]].head(10)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('indonesian'))

def remove_stopwords(text):
    tokens = text.split()
    filtered = [word for word in tokens if word not in stop_words]
    return " ".join(filtered)

df["clean_text_nostop"] = df["clean_text"].apply(remove_stopwords)
df[["clean_text", "clean_text_nostop"]].head()
```

```python
# Buat folder jika belum ada
output_dir = "../data/processed"
os.makedirs(output_dir, exist_ok=True)

# Pilih kolom yang disimpan
df_cleaned = df[["clean_text"]]  # atau ["clean_text_nostop"]

# Simpan ke CSV
df_cleaned.to_csv(os.path.join(output_dir, "gojek_cleaned.csv"), index=False)

print("✅ Data berhasil disimpan ke data/processed/gojek_cleaned.csv")
```
